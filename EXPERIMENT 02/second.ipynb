```python
# Experiment 2 - Introdiction to Numpy and Pandas 
# PROBLEM :- Given a list of numbers [2,4,6,8,10,12,14,16], perform the following tasks using Numpy functions;

import numpy as np
import pandas as pd

# Given list
list1 = [2, 4, 6, 8, 10, 12, 14, 16]

# 1. Create a NumPy array
N_array = np.array(list1)
print("List in Python:", list1)
print("Numpy Array in Python:", N_array)

# 2. Mean, Median, Std Deviation
print("Mean:", np.mean(N_array))
print("Median:", np.median(N_array))
print("Standard Deviation:", np.std(N_array))

# 3. Reshape to 2x4 matrix
reshaped_array = N_array.reshape(2, 4)
print("Reshaped Array (2x4):\n", reshaped_array)

# 4. Element-wise square root
sqrt_array = np.sqrt(N_array)
print("Square Root of Each Element:\n", sqrt_array)

# 5. Index of maximum element
max_index = np.argmax(N_array)
print("Index of Maximum Element:", max_index)

# 6. Elements greater than 8
greater_than_8 = N_array[N_array > 8]
print("Elements Greater Than 8:", greater_than_8)
```

```python
# 2. 2.2 Pandas:
# Consider the following data about employees in a company:
# Name	Age	Department	Salary
# Alice	28	HR		50000
# Bob	22	Engineering	60000
# Charlie	35	HR		55000
# David	30	Engineering	62000
# Eve	25	Finance		45000

import pandas as pd

# 1. Create the DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [28, 22, 35, 30, 25],
    'Department': ['HR', 'Engineering', 'HR', 'Engineering', 'Finance'],
    'Salary': [50000, 60000, 55000, 62000, 45000]
}

df = pd.DataFrame(data)

# 2. Display basic information about the DataFrame
print("Basic Info:\n")
print(df.info())

# 3. Calculate the average salary
avg_salary = df['Salary'].mean()
print(f"\nAverage Salary: {avg_salary}")

# 4. Maximum age of employees in Engineering
max_age_engineering = df[df['Department'] == 'Engineering']['Age'].max()
print(f"\nMaximum Age in Engineering: {max_age_engineering}")

# 5. Names of employees with salary > 50000
high_salary_names = df[df['Salary'] > 50000]['Name']
print("\nEmployees with salary > 50000:")
print(high_salary_names.to_string(index=False))

# 6. Add a new employee: Frank
new_employee = pd.DataFrame({
    'Name': ['Frank'],
    'Age': [40],
    'Department': ['Finance'],
    'Salary': [55000]
})
df = pd.concat([df, new_employee], ignore_index=True)

# 7. Remove the employee named "Eve"
df = df[df['Name'] != 'Eve']

# 8. Final DataFrame
print("\nFinal DataFrame:")
print(df)
```
